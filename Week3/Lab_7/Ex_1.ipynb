```python
import numpy as np  
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-10,10,500)
y1 = np.maximum(0, 1 - x) 
y2 = np.log(1 + np.exp(-x)) 
y3 = (np.maximum(0, 1 - x)) ** 2
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y1 = max(0, 1 - x)', color='blue', linewidth=2)
plt.plot(x, y2, label='y2 = log(1 + exp(-x))', color='green', linewidth=2)
plt.plot(x, y3, label='y3 = (max(0, 1 - x))^2', color='red', linewidth=2)
```
