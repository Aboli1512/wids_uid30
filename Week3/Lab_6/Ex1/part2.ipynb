```python
import numpy as np
from sklearn.datasets import load_digits
digits = load_digits()
 #check the shape of digits data
print(digits.data.shape)
 #check the shape of digits target
print(digits.target.shape)
 #let us use the linear regression used in the previous lab
 #N = digits.data.shape[0] #Number of data points
 #d = digits.data.shape[1] #Dimension of data points
A = digits.data
 #In the following code, we create a NX1 vector of target labels
y = 1.0*np.ones([A.shape[0],1])
for i in range(digits.target.shape[0]):
   y[i] = digits.target[i]
print(y.shape)
```

```python
def f(x,y,A) :
    return 0.5*((A@x - y).T @ (A@x - y))

```

```python
def grad_f(x,y,A) :
    return A.T @ (A@x - y)
```

```python
def L2Norm(x,y,A) :
    return np.linalg.norm(grad_f(x,y,A))
```

```python
def alpha_cal(p,A,y,x) :
    numerator = 0.5 * ((p.T @ A.T @ y) + (y.T @ A @ p) - (x.T @ A.T @ A @ p) - (p.T @ A.T @ A @ x))
    denominator = p.T @ A.T @ A @ p
    return numerator / denominator
```

```python
def gradient_descent(x_start,y,A) :

    x = x_start
    count = 0
    norm = L2Norm(x,y,A)
    TLevel = 10**(-15)
    H = np.identity(A.shape[1])

    while(norm > TLevel) :
        
        p = -H @ grad_f(x,y,A)
        alpha = alpha_cal(p,A,y,x)

        s = alpha * p
        x_present = x
        x = x + s

        yk = grad_f(x,y,A) - grad_f(x_present,y,A)

        skT_yk = s.T @ yk

        if np.abs(skT_yk) < 1e-10:  # Avoid division by zero
            break

        H = H + (skT_yk + yk.T @ H @ yk) * (s @ s.T) / (skT_yk ** 2) - (H @ yk @ s.T + s @ yk.T @ H) / skT_yk
        norm = L2Norm(x,y,A)
        count = count + 1
    
    print(f"Iterations: {count}")
    print(x,f(x,y, A))
    
    return x,y,f(x,y,A)
```

```python
x0 = np.zeros((64,1))
x_dash = gradient_descent(x0,y,A)

print(x_dash)
```

```python
def alpha_cal(p, A, y, x, lambdaa):
    # Compute the gradient of the regularized function
    grad = A.T @ (A @ x - y) + lambdaa * x
    
    # Compute the numerator using the updated formula for regularized OLS
    numerator = 0.5 * ((p.T @ A.T @ y) + (y.T @ A @ p) - (x.T @ A.T @ A @ p) - (p.T @ A.T @ A @ x))
    
    # Add the regularization part to the denominator
    denominator = p.T @ (A.T @ A + lambdaa * np.identity(A.shape[1])) @ p
    
    # Return the step size alpha for regularized OLS
    return numerator / denominator
```
